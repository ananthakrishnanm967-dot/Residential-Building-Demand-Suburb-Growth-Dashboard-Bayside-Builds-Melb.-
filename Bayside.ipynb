```python
melbourne_pop["population_increase"] = (
    melbourne_pop["population_2024"] -
    melbourne_pop["population_2023"]
)

melbourne_pop["population_growth_rate"] = (
    melbourne_pop["population_increase"] /
    melbourne_pop["population_2023"]
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

melbourne_pop[[
    "growth_score",
    "size_score",
    "increase_score"
]] = scaler.fit_transform(
    melbourne_pop[[
        "population_growth_rate",
        "population_2024",
        "population_increase"
    ]]
)
```

```python
melbourne_pop["development_opportunity_score"] = (
      0.5 * melbourne_pop["growth_score"]
    + 0.3 * melbourne_pop["size_score"]
    + 0.2 * melbourne_pop["increase_score"]
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

viz = melbourne_pop.rename(columns={
    "suburb": "Suburb",
    "population_2024": "Current Population",
    "population_increase": "New Residents",
    "population_growth_rate": "Growth Rate",
    "development_opportunity_score": "Development Score"
})

viz["Growth (%)"] = viz["Growth Rate"] * 100
```

```python
top10 = viz.sort_values(
    "Development Score",
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

bars = plt.barh(
    top10["Suburb"],
    top10["Development Score"],
    color="darkgreen"
)

plt.title("Best Melbourne Suburbs for New Housing Projects")
plt.xlabel("Overall Development Opportunity Score")

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.01,
             bar.get_y() + bar.get_height()/2,
             f"{width:.2f}",
             va="center")

plt.gca().invert_yaxis()
plt.show()
```

```python
top10_people = viz.sort_values(
    "New Residents",
    ascending=False
).head(10)

plt.figure(figsize=(10,6))

bars = plt.bar(
    top10_people["Suburb"],
    top10_people["New Residents"],
    color="orange"
)

plt.title("Suburbs with the Largest Population Increase")
plt.ylabel("New Residents (2023–2024)")

plt.xticks(rotation=40)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        height,
        f"{int(height)}",
        ha="center",
        va="bottom"
    )

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create readable growth categories
bins = [0, 2, 5, 10, 20, 40]
labels = [
    "Very Low Growth (<2%)",
    "Low Growth (2–5%)",
    "Moderate Growth (5–10%)",
    "High Growth (10–20%)",
    "Rapid Growth (>20%)"
]

viz["Growth Category"] = pd.cut(
    viz["Growth (%)"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Count suburbs per category
category_counts = viz["Growth Category"].value_counts().sort_index()

plt.figure(figsize=(9,5))

bars = plt.bar(
    category_counts.index,
    category_counts.values,
    color=["gray","skyblue","gold","orange","red"]
)

plt.title("How Fast Melbourne Suburbs Are Growing", fontsize=14)
plt.ylabel("Number of Suburbs")
plt.xlabel("Population Growth Category")

plt.xticks(rotation=20)

# Add labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{int(height)}",
        ha="center",
        va="bottom"
    )

plt.tight_layout()
plt.show()
```
